First plot data points on a scatter plot chart to see if linear regression model can be applied. If yes, then build a model that can answer following questions,

1) Predict price of a mercedez benz that is 4 yr old with mileage 45000

2) Predict price of a BMW X5 that is 7 yr old with mileage 86000

3) Tell me the score (accuracy) of your model. (Hint: use LinearRegression().score())

```python
import pandas as pd
```

```python
df = pd.read_csv('car.csv')
df
```

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
merged  = pd.concat([df, dummies], axis='columns')
merged
```

```python
final = merged.drop(['Car Model', 'Mercedez Benz C class'], axis='columns')
final
```

```python
pd.plotting.scatter_matrix(final)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X= final.drop('Sell Price($)', axis='columns')
X
```

```python
y = final['Sell Price($)']
y
```

```python
model.fit(X,y)
```

1) Predict price of a mercedez benz that is 4 yr old with mileage 45000

```python
model.predict([[45000, 4, 0,0]])
```

2) Predict price of a BMW X5 that is 7 yr old with mileage 86000

```python
model.predict([[86000, 7, 0,1]])
```

3. Tell me the score (accuracy) of your model.

```python
model.score(X,y)
```
